Un empresario posee una flotilla de 100 vehículos que utiliza como el servicio de taxi en una comunidad, como es natural a veces un auto puede necesitar un servicio de mantenimiento o bien es posible que presente pequeñas fallas y pueda ser necesario una reparación bastante simple a estos sucesos los denominaremos reparaciones menores y en otras ocasiones es posible que un auto presente daños fuertes en el motor o tenga un accidente cuya reparación necesite más días en un taller a este tipo de eventos se les denomina reparaciones mayores.



Determina la  probabilidad de que el día de mañana un auto se encuentre en buen estado dado que el día de hoy se encuentra en buen estado es 0.97 la probabilidad de que el día de mañana un auto necesite reparaciones menores dado que hoy se encuentra en buen estado es 0.02 y la probabilidad de que el día de mañana un auto necesite reparaciones mayores dado que el día de hoy se encuentra en buen estado es 0.01



Buen estado(A)
* A a A = 0.97
* A a B = 0.02
* A a C = 0.01

Reparacion Menor(B)
* B a A = 0.6
* B a B = 0.3
* B a C = 0.1

Reparacion Mayor(C)
* C a A = 0.2
* C a B = 0.1
* C a C = 0.7

```python
import numpy as np
import pandas as pd
```

```python
a_a = 0.97
a_b = 0.02
a_c = 0.01
b_a = 0.60
b_b = 0.30
b_c = 0.10
c_a = 0.2
c_b = 0.1
c_c = 0.7

M1 = np.array([[a_a, a_b, a_c], [b_a, b_b, b_c], [c_a, c_b, c_c]])

print(M1)
```

```python
def SumaFilas(M: np.array) :
    for i in range(M.shape[0]):
        suma = 0
        for j in range(M.shape[1]):
            suma += M[i][j]
            #print(M[i][j])
            suma_redondeada = round(suma, 10)
        print(f"La suma de la fila {i} es {suma_redondeada}")
#si en alguna llega a ser diferente a 1 algo estaria mal
SumaFilas(M1)
```

```python
def CoincidenColumnas(M: np.array) -> bool:
    for i in range(M.shape[1]):  

        primera_valor = M[0][i]  
        for j in range(1, M.shape[0]):  
            if round(M[j][i], 10) != round(primera_valor, 10):
                return False  
    return True  

CoincidenColumnas(M1)
```

```python
def MultiplicaMatrices(A: np.array, B: np.array) -> np.array:
    
    C = np.zeros((A.shape[0], B.shape[1]))
    for i in range(A.shape[0]):
        for j in range(B.shape[1]):
            for k in range(A.shape[1]):
                C[i][j] += A[i][k] * B[k][j]
    return C

M2=MultiplicaMatrices(M1, M1)
print(M2)
```

```python
Mn=MultiplicaMatrices(M1, M1)
contador = 2
titulos = ["Buen Estado", "Reparacion Menor", "Reparacion Mayor"]
M1DF = pd.DataFrame(M1, columns=titulos, index=titulos)
MnDF = pd.DataFrame(Mn, columns=titulos, index=titulos)
print(f"Matriz M1 \n{M1DF}")
SumaFilas(M1)

print(f"\n\nMatriz M2 \n{MnDF}")
SumaFilas(Mn)
0

while not CoincidenColumnas(Mn):
    contador += 1
    print(f"\n\nMatriz M{contador}")
    Mn = MultiplicaMatrices(Mn, M1)
    MnDF = pd.DataFrame(Mn, columns=titulos, index=titulos)
    print(MnDF)
    SumaFilas(Mn)
```
