Teniendo en cuenta una lista de 3 calificaciones para cada alumno, mostrar en un gráfico las calificaciones medias (promedio del alumno) usando matplotlib, en un gráfico de barras etiquete los datos en cada uno de los alumnos en las barras. Muestre el promedio numérico en cada barra.

Ejemplo del array de datos

nombres = ["Pedro","Maria","Jose","Karina"]

calificaciones = [[2,4.5,9],[5,3,10],[7.5,8.3,9],[4,6.5,7]]

Nota: se recomienda utilizar las librerías numpy o pandas de python

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nombres = np.array(['Pedro','Maria','Jose','Karina'])
calificaciones = np.array([[2,4.5,9],[5,3,10],[7.5,8.3,9],[4,6.5,7]])
```

```python

promedios = [np.mean(cal) for cal in calificaciones]
promedios.sort()

bars = plt.bar(nombres, promedios, color='skyblue')

for bar, promedio in zip(bars, promedios):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(promedio, 2), ha='center', va='bottom')

plt.title('Promedio de Calificaciones por Alumno')
plt.xlabel('Alumnos')
plt.ylabel('Promedio de Calificaciones')

plt.show()
```
